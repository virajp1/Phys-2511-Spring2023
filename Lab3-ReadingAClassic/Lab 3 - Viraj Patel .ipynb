Counts the Occurances of Characters in Frankenstein

```python
name=input("What charcter are you searching for?:  ")
with open("frankenstein.txt", "r") as file:
    text = file.read()
chapters = text.split("Chapter")
character = name
chapter_num = 1
for chapter in chapters:
    count = chapter.count(name)
    print("Chapter\t" + str(chapter_num) + ":\t" + str(name) + "\toccurs\t" + str(count))
    chapter_num += 1
```

Graph showing the occurances of different characters

```python
import matplotlib 
import matplotlib.pyplot as plt

def bar(x_axis, y_axis):
    plt.bar(x_axis, y_axis)
    plt.xlabel("Characters")
    plt.ylabel("Occurrence")
    plt.title("Bar Graph of Character Occurrence")
    plt.show()

x_axis = ["Victor", "Creature", "Agatha", "Caroline", "De Lacey", "Elizabeth", "Ernest", "Felix", "Henry", "Justine", "William"]
y_axis = [28, 69, 22, 3, 9, 92, 13, 50, 26, 55, 25]

bar(x_axis, y_axis)
```
