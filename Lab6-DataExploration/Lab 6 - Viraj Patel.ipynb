```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
import csv

dates=[]
gdp_values=[]

with open ('Downloads/GDPC1.csv','r') as csvfile:
    reader=csv.reader(csvfile)
    next(reader)
    for row in reader:
        dates.append(row[0])
        gdp_values.append(float(row[1]))
        
fig,ax=plt.subplots()


    
plt.plot(dates, gdp_values)
plt.xlabel("Date (Quarterly)")
plt.ylabel("GDP (U.S. Dollar)")
plt.title("U.S. Real Gross Domestic Product Over Time (1947-2022)")
ax.set_xticks(np.arange(0,len(dates), 10))
ax.set_xticklabels(dates[::10], rotation=90)
plt.show()
    

```

The line graph depicts the U.S. GDP as a function of time. The x-axis shows the dates that starting from January 1st 1947 through January 1 2022. The increments for the dates is in each fiscal quarter. The y-axis shows the increasing of the GDP in U.S. Dollars. 

Data Citation: 

U.S. Bureau of Economic Analysis, Real Gross Domestic Product [GDPC1], retrieved from FRED, Federal Reserve 
    Bank of St. Louis; https://fred.stlouisfed.org/series/GDPC1, February 28, 2023.
